```
train_data = train_data.loc[(train_data['分类']=='体育')|(train_data['分类']=='财经')|(train_data['分类']=='科技')]
train_data.head()
```

```
# 载入停用词
stopwords = set()
with open('/content/drive/MyDrive/Colab Notebooks/data/stopwords.txt', 'r') as infile:
  for line in infile:
    line = line.rstrip('\n')
    if line:
      stopwords.add(line.lower())
```

## 2. 计算tfidf

```
import jieba
from sklearn.feature_extraction.text import TfidfVectorizer

# min_df 去掉df值最小的词
# max_df 去掉df值很大的词
# norm 表示对TF-IDF矩阵的每一行使用l2范数正则化(变成单位向量 模长为1)   l1范数就是绝对值的和   max正则化---最大值
# use_idf: 表示在TF矩阵的基础上计算IDF，并相乘得到TF-IDF
# smooth_idf: 表示计算IDF时，分子上的总文档数+1
# sublinear_tf: 表示使用 1+log(tf)替换原来的tf
tfidf = TfidfVectorizer(tokenizer=jieba.lcut, stop_words=stopwords, min_df=0.1, max_df=0.6, norm='l2', use_idf=True, smooth_idf=False, sublinear_tf=False)
x = tfidf.fit_transform(train_data[u'内容'])
```

```
#返回的是稀疏矩阵
x.toarray()
```

```
print(u'词表大小：{}'.format(len(tfidf.vocabulary_)))
```

## 3. 训练模型

```
# 编码目标变量 
from sklearn.preprocessing import LabelEncoder
y_encoder = LabelEncoder()
y = y_encoder.fit_transform(train_data[u'分类'])
y

x = tfidf.transform(train_data[u'内容'])
x
```

```
# 划分训练集
from sklearn.model_selection import train_test_split
train_idx, valid_idx = train_test_split(range(len(y)), test_size=0.3, stratify=y)
train_x = x[train_idx, :]
train_y = y[train_idx]
valid_x = x[valid_idx, :]
valid_y = y[valid_idx]
```

```
# 训练贝叶斯模型

from sklearn import naive_bayes, metrics


model_nb = naive_bayes.MultinomialNB()
model_nb.fit(train_x, train_y)
pred_y = model_nb.predict(valid_x)
micro_f1 = metrics.f1_score(valid_y, pred_y, average='micro')
macro_f1 = metrics.f1_score(valid_y, pred_y, average='macro')

print('micro f1:',micro_f1)
print('macro f1:',macro_f1)
```

## 4. 评价指标

```
from sklearn.metrics import precision_recall_fscore_support


precision_recall_fscore_support(valid_y, pred_y)
```

```
# 计算各项评价指标
def eval_model(y_true, y_pred, labels):
    # 计算每个分类的Precision, Recall, f1, support
    p, r, f1, s = precision_recall_fscore_support(y_true, y_pred)
    # 计算总体的平均Precision, Recall, f1, support
    tot_p = np.average(p, weights=s)
    tot_r = np.average(r, weights=s)
    tot_f1 = np.average(f1, weights=s)
    tot_s = np.sum(s)
    res1 = pd.DataFrame({
        u'Label': labels,
        u'Precision': p,
        u'Recall': r,
        u'F1': f1,
        u'Support': s
    })
    res2 = pd.DataFrame({
        u'Label': [u'总体'],
        u'Precision': [tot_p],
        u'Recall': [tot_r],
        u'F1': [tot_f1],
        u'Support': [tot_s]
    })
    res2.index = [4]
    res = pd.concat([res1, res2])
    return res[[u'Label', u'Precision', u'Recall', u'F1', u'Support']]
```

```
eval_model(valid_y, pred_y, y_encoder.classes_)
```

```
precision_recall_fscore_support(valid_y, pred_y,  average='macro')
```
