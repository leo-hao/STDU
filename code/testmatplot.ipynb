```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df=pd.read_excel('data.xlsx')
data=df.loc[:,['OXY','time','age']].values
print("读取OXY, time, age 列的数据：\n{0}".format(data))
```

```python
df.plot(x='time', y='OXY', kind='scatter')
```

```python
df.plot(x='age', y='OXY', kind='scatter')
```

```python
pd.plotting.scatter_matrix(df, figsize=(10,10), alpha=0.6, diagonal='kde')
```

```python
del df['序号']
data=df.loc[[0,1,20,21]].values
plt.plot(data)
print('读取1，2, 21, 22 行的数据：\n{0}'.format(data))
```

```python
print(data[1])
```

```python
x = ['age', 'weight', 'time', 'spulse', 'rpulse' , 'mpulse',' OXY']
```

```python
y0 = data[0]
y1 = data[1]
y2 = data[2]
y3 = data[3]
plt.plot(x, y0, marker='*', ms=10, label="1")
plt.plot(x, y1, marker='*', ms=10, label="2")
plt.plot(x, y2, marker='*', ms=10, label="21")
plt.plot(x, y3, marker='*', ms=10, label="22")
```

```python
name = ['age', 'weight', 'time', 'spulse', 'rpulse' , 'mpulse',' OXY']
theta = np.linspace(0,2*np.pi,len(name),endpoint=False)    #将圆根据标签的个数等比分
plt.figure(figsize=(8,4))


ax1= plt.subplot(111, projection='polar')
ax1.set_title('radar map\n')  
ax1.set_rlim(0,12)
ax1.set_thetagrids(theta*180/np.pi,name)         #替换标签
ax1.set_ylim(0,110)                          #设置极轴的区间
ax1.set_theta_zero_location('N')         #设置极轴方向

ax1.plot(theta,y0,'.--',label='data0')
ax1.fill(theta,y0,alpha=0.2)
ax1.plot(theta,y1,'.--',label='data1')
ax1.fill(theta,y1,alpha=0.2)
ax1.plot(theta,y2,'.--',label='data2')
ax1.fill(theta,y2,alpha=0.2)
ax1.plot(theta,y3,'.--',label='data3')
ax1.fill(theta,y3,alpha=0.2)
```
